```python
import string
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
```

```python
import nltk
nltk.download('punkt')
```

```python
text = open('Documents/read.txt',encoding='utf-8').read()
```

```python
text
```

```python
#convert into lower case
text_lower=text.lower()
```

```python
text_lower
```

```python
#remove punctuation
cleaned_text = text_lower.translate(str.maketrans('','',string.punctuation))
cleaned_text
```

```python
# splitting text into words
tokenized_words = word_tokenize(cleaned_text,'english')
tokenized_words
```

```python

# Removing stop words from the tokenized words list
from nltk.corpus import stopwords

# Load the stopwords for English
stop_words = set(stopwords.words('english'))

final_words = []
for word in tokenized_words:
    if word.lower() not in stop_words:
        final_words.append(word)
```

```python
nltk.download('stopwords')
```

```python
final_words
```

```python
# 1) Check if the word in the final word list is also present in emotion.txt
#  - open the emotion file
#  - Loop through each line and clear it
#  - Extract the word and emotion using split

# 2) If word is present -> Add the emotion to emotion_list
# 3) Finally count each emotion in the emotion list

emotion_list = []
with open('Documents/emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n','').replace(',','').replace("'",'').strip()
        word, emotion = clear_line.split(':')
        print('Word:'+word+' '+ 'Emotion:'+emotion)
        
        if word in final_words:
            emotion_list.append(emotion)
            
```

```python
print(emotion_list)
```

```python

w=Counter(emotion_list)
w
```

```python
# Plotting the emotions on the graph

fig, axl = plt.subplots()
axl.bar(w.keys(),w.values())
fig.autofmt_xdate()
plt.savefig('graph.png')
plt.show()
```
